In this notebook, I am trying to fetch image from unsplash for experimentation 

```python
# Import Libs

# General
import numpy as np
import pandas as pd

# Import Cv2
import cv2

# Import Requests
import requests

# Pillow
from PIL import Image
```

The first step would be to go into unsplash & find the image that you are looking from.

Every Image would have a image id associated from it ex: IF9TK5Uy-KI

This would be in the last when we open the image.

Next create the web address using the format: *https://source.unsplash.com/<Image-ID>*

example: https://source.unsplash.com/IF9TK5Uy-KI

Once the URL is read we can use the requests library to get the image

```python
# Read the Image
img = Image.open(requests.get('https://source.unsplash.com/IF9TK5Uy-KI', stream=True).raw)
img
```
